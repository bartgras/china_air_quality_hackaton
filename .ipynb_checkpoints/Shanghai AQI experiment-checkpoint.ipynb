### We've decided to switch to other AQI data source and only get data for Shanghai

#### Fltering in terminal

less aqi.csv | grep shanghai

less aqi.csv | grep 310000 > shanghai_stations.csv

```python
header_names = ['ID','division','areaName','value','pollutant','recordDate','_fetchDate','source']
shangai_raw = pd.read_csv('shanghai_stations.csv', names=header_names)
shangai_raw.head()
```

```python
shangai_raw_cleaned = shangai_raw.ix[:,['ID','division','value','pollutant','recordDate']]
shangai_raw_cleaned.head()
```

```python
shangai_raw_cleaned.shape
```

```python
shangai_raw_cleaned.groupby('pollutant').count()
```

```python
filtered_shanghai = shangai_raw_cleaned[shangai_raw_cleaned.pollutant == 1]
filtered_shanghai.head()
```

```python
filtered_shanghai.shape
```

```python
import datetime
# datetime.datetime.fromtimestamp(d)

filtered_shanghai['date'] = filtered_shanghai['recordDate'].map(lambda d: datetime.datetime.fromtimestamp(d))
```

```python
filtered_shanghai.head(2)
```

```python
filtered_shanghai.tail(2)
```

### So we have Shanghai AQI data FROM: 2000-06-05 to 2013-01-14

```python
%matplotlib inline

filtered_shanghai.plot(x='date',y='value',figsize=(15,10))
```

### How about 1 year: 2012-01 - 2013-01

```python
filtered_shanghai[filtered_shanghai['date'] > '2012-01-01'].plot(x='date',y='value',figsize=(15,10))
```
